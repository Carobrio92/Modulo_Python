```python
import pandas as pd
import numpy as np
```

```python
# SE ACCEDIO A LA BASE IMPORTANDO PAQUETES PANDAS Y NUMPY
# LUEGO PD.READ.CSV IMPRIMIR LA BASE FELICIDAD
base_felicidad=pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
```

```python
base_felicidad
```

```python
# SE UTILIZO LA INDEXACION BOOLEANA UTILIZANDO FILTRO DEL MAX Y MIN
data_bool=(base_felicidad.anio==max(base_felicidad.anio)) | (base_felicidad.anio==min(base_felicidad.anio))
```

```python
data_observacion=base_felicidad[data_bool]
data_observacion[data_observacion['pais']=='Francia']
```

```python
# MEDIDA DE TENDENCCIA CENTRAL PARA EL PRIMER ANO A NIVEL MUNDIAL
promedio_gen_primer_ano=data_observacion[data_observacion['anio']==min(data_observacion['anio'])].escalera_vida.mean()
promedio_gen_primer_ano
```

```python
# MEDIDA DE TENDENCIA CENTRAL PARA EL ULTIMO ANO NIVEL MUNDIAL
promedio_gen_ultimo_ano=data_observacion[data_observacion['anio']==max(data_observacion['anio'])].escalera_vida.mean()
promedio_gen_ultimo_ano
```

```python
# DESVIACION ESTANDAR PARA EL PRIMER ANO A NIVEL GLOBAL
promedio_disp_primer_ano=data_observacion[data_observacion['anio']==min(data_observacion['anio'])].escalera_vida.std()
promedio_disp_primer_ano
```

```python
# DESVIACION ESTANDAR ULTIMO ANO A NIVEL GLOBAL
promedio_disp_ultimo_ano=data_observacion[data_observacion['anio']==max(data_observacion['anio'])].escalera_vida.std()
promedio_disp_ultimo_ano
```

```python
# CON LA INDEXACION BOOLEANA TAMBIEN SE FILTRO PARA LOS PAISES CANADA Y FRANCIA
data_paises = data_observacion[data_observacion['pais'].isin(['Canadá','Francia'])]
```
