```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```


**1.ACCEDER A LOS DATOS**

**Base De Datos Partidos de la Copa de fútbol masculina**

```python
partidos_copa_mundial= pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt', delimiter = '\t')
partidos_copa_mundial.head()
```

**1.1 PARA LOS PARTIDOS DEL MUNDIAL SE ESCOGIO DOS PAISES ESPANA Y ALEMANIA PARA COMPARAR CON LOS EQUIPOS  1 Y 2**

```python

partidos_paises = partidos_copa_mundial[(partidos_copa_mundial.anfitrion == 'España') | (partidos_copa_mundial.anfitrion == 'Alemania')]
partidos_paises
```

**2.CREAR GRAFICOS COMPAREN LOS DOS PAISES**

**2.1 Gráfico 1.Los datos de los paises aparezcan en un mismo cuadro**

```python
fig=plt.figure()
eje=fig.add_subplot(1,1,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'fecha',y = 'equipo_1_final',ax = eje,linestyle = '-.')
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'fecha',y = 'equipo_2_final',ax = eje,linestyle = '--')
plt.xlabel('España')
plt.ylabel('equipo_2_final')
plt.title('Partidos España')
# PARA GUARDAR LA FIGURA CON PLT.SAVEFIG
plt.savefig('mi_figura.tiff')
```

```python
fig=plt.figure()
eje=fig.add_subplot(1,1,1)
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'fecha',y = 'equipo_1_final',ax = eje,linestyle = '--')
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'fecha',y = 'equipo_2_final',ax = eje,linestyle = '-.')
plt.xlabel('Alemania')
plt.ylabel('equipo_2_final')
plt.title('Partidos Alemania')
plt.savefig('mi_figura.tiff')
```

```python
fig=plt.figure()
eje=fig.add_subplot(1,1,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot.line(x = 'fecha',y = 'equipo_1_final',ax = eje,linestyle = 'dotted', color = 'red')
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot.line(x = 'fecha',y = 'equipo_1_final',ax = eje,linestyle = '--', color = 'green')
plt.xlabel('fecha')
plt.ylabel('Espana_Alemania equipo_final_1')
plt.title('Comparación los paises España y Alemania partidos final por equipos #1')
plt.savefig('mi_figura.tiff')
```

```python
fig=plt.figure()
eje=fig.add_subplot(1,1,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot.line(x = 'fecha',y = 'equipo_2_final',ax = eje,linestyle = 'dotted', color = 'blue')
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot.line(x = 'fecha',y = 'equipo_2_final',ax = eje,linestyle = '-.', color = 'orange')
plt.xlabel('fecha')
plt.ylabel('Paises Espana_Alemania equipo_final_2')
plt.title('Comparación los paises España y Alemania partidos final #2')
plt.savefig('mi_figura.tiff')
```

**2.2 Gráfico 2.Los datos de los paises aparezcan en un subgráfico diferente**

**Se comparan ambos paises por equipos final 1, solo muestra un solo ano por fecha**

```python
fig=plt.figure()
eje=fig.add_subplot(1,2,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'fecha',y = 'equipo_1_final',ax = eje,linestyle = '--', color = 'green')
#Agrega el segundo eje
eje2 = fig.add_subplot(1, 2, 2, sharey = eje)
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'fecha',y = 'equipo_1_final',ax = eje2,linestyle= 'dotted', color = 'blue')
eje.set_xlabel('España')
eje2.set_xlabel('Alemania')
eje.set_ylabel('equipo_1_final')
plt.title('Comparación  gráficos por paises España y Alemania partidos final 1')
plt.savefig('mi_figura.tiff')
```

**Comparar dos paises por equipo final 2**

```python
fig=plt.figure()
eje=fig.add_subplot(1,2,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'fecha',y = 'equipo_2_final',ax = eje,linestyle = '-', color = 'red')
#Agrega el segundo eje
eje2 = fig.add_subplot(1, 2, 2, sharey = eje)
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'fecha',y = 'equipo_2_final',ax = eje2,linestyle= '-', color = 'orange')
eje.set_xlabel('España')
eje2.set_xlabel('Alemania')
eje.set_ylabel('equipo_2_final')
plt.title('Comparación  gráficos de España y Alemania partidos final 2')
plt.savefig('mi_figura.tiff')
```

**TAMBIEN SE COMPARO LOS EQUIPO FINAL 1 Y 2 CON LOS EQUIPO 1 Y 2 EN PAISES**

```python
fig=plt.figure()
eje=fig.add_subplot(2,2,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'equipo_1',y = 'equipo_1_final',ax = eje,linestyle = ':', color = 'brown')
eje2 = fig.add_subplot(2, 2, 2, sharey = eje)
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'equipo_1',y = 'equipo_1_final',ax = eje2,linestyle= '--', color = 'blue')
eje.set_xlabel('España')
eje2.set_xlabel('Alemania')
eje.set_ylabel('equipo_2_final')
plt.title('Comparación  de España y Alemania entre equipos 1 y equipo final por pais')
```

```python
fig=plt.figure()
eje=fig.add_subplot(2,2,1)
partidos_paises[partidos_paises['anfitrion']=='España'].plot(x = 'equipo_2',y = 'equipo_2_final',ax = eje,linestyle = '-', color = 'purple')
eje2 = fig.add_subplot(2, 2, 2, sharey = eje)
partidos_paises[partidos_paises['anfitrion']=='Alemania'].plot(x = 'equipo_2',y = 'equipo_2_final',ax = eje2,linestyle= '--', color = 'gray')
eje.set_xlabel('España')
eje2.set_xlabel('Alemania')
eje.set_ylabel('equipo_2_final')
plt.title('Comparación de España y Alemania entre equipos 2 por pais')
```
